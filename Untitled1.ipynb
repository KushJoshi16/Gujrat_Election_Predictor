```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
text="hello @harshitbajeta i am #first"
doc=nlp(text)
for word in doc.ents:
        print("hello")
        print(word.text,word.label_)
```

```python
doc = nlp(u'Microsoft is trying to buy harshitbajeta France based startup at $7 Million')
for token in doc:
  print(token.text)
```

```python
for token in doc:
    print(token, token.pos_)
```

```python
for word in doc.ents:
    print(word.text,word.label_,str(spacy.explain(word.label_)))
```
