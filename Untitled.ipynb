```python
 # removal of 
def clean(text):
    text= re.sub(r'https?://[^ ]+', '', text)  # links
    text= re.sub(r'@[^ ]+', '', text)           # username
    text=re.sub(r'#', '', text)                    # hashtag
    #text=emoji.demojize(text)                # emoji to text
    text=re.sub(r'[^A-Za-z ]', '', text)      
    return text
df1['Tweet Text']=df1['Tweet Text'].apply(lambda x: clean(x))
```

```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
from nltk.stem import PorterStemmer
porter = PorterStemmer()
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
def stemSentence(sentence):
    token_words=word_tokenize(sentence)
    token_words
    stem_sentence=[]
    for word in token_words:
        stem_sentence.append(porter.stem(word))
        stem_sentence.append(" ")
    return "".join(stem_sentence)
df2['Tweet Text'] = df2['Tweet Text'].apply(lambda x : stemSentence(x))
df2
```

```python
b=set()
def ner1(text):
    doc=nlp(text)
    for word in doc.ents:
        b.add(word.text)
for i in df2['Tweet Text']:
    ner1(i)
```

```python
s = list(b)
df3= pd.DataFrame(s,columns =['NER'])
df3.to_csv('NER4.csv')
df3
```
