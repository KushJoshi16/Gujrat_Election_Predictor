```python
import pandas as pd
import re
```

```python
df=pd.read_csv("Tweets1.csv")
df
```

```python
df=df.drop(['Tweet Scraped','Username','Location','Hashtags Used','Description'],axis=1)
```

```python
df.to_csv('data.csv')
```

```python
df['Tweet Text']=df['Tweet Text'].str.lower()
```

```python
def clean(text):
    text= re.sub(r'https?://[^ ]+', '', text)  # links
    text=re.sub(r'[^A-Za-z ]', '', text)      
    return text
df['Tweet Text']=df['Tweet Text'].apply(lambda x: clean(x))
```

```python
df
```

```python
import spacy
from spacy import displacy
nlp = spacy.load("en_core_web_sm")
```

```python
a=set()
def ner(text):
    doc=nlp(text)
    for word in doc.ents:
        a.add(word.text)
       # print(word.text,word.label_,str(spacy.explain(word.label_)))
       #print(a)
        
for i in df['Tweet Text']:
    ner(i)
s = list(a)
df1= pd.DataFrame(s,columns =['NER'])
df1.to_csv('NER2.csv')
df1
```

```python
df2=df
import nltk
# nltk.download('wordnet')
# nltk.download('omw-1.4')
```

```python
from nltk.stem import PorterStemmer
porter = PorterStemmer()
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
def stemSentence(sentence):
    token_words=word_tokenize(sentence)
    token_words
    stem_sentence=[]
    for word in token_words:
        stem_sentence.append(porter.stem(word))
        stem_sentence.append(" ")
    return "".join(stem_sentence)
df2['Tweet Text'] = df2['Tweet Text'].apply(lambda x : stemSentence(x))
df2
```

```python
b=set()
def ner1(text):
    doc=nlp(text)
    for word in doc.ents:
        b.add(word.text)
for i in df2['Tweet Text']:
    ner1(i)
s = list(b)
df3= pd.DataFrame(s,columns =['NER'])
df3.to_csv('NER5.csv')
df3
```
